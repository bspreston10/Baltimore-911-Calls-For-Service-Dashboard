```python
# import packages
import pandas as pd 
import numpy as np 
```

# Calls for service 2024

```python
data_2024 = pd.read_csv('911_Calls_For_Services_2024.csv')
```

```python
data_2024.head()
```

```python
data_2024.columns
```

## Dropping columns

```python
data_2024 = data_2024.drop(columns=['incidentLocation', 'PolicePost', 'CouncilDistrict', 'SheriffDistricts', 'Community_Statistical_Areas', 
                        'Census_Tracts', 'VRIZones', 'NeedsSync', 'IsDeleted', 'HashedRecord', 'ESRI_OID'])
```

## Fixing data types

```python
data_2024.dtypes
```

- callDateTime to datetime
- ZIPCode to string

```python
data_2024['callDateTime'] = pd.to_datetime(data_2024['callDateTime'])
```

```python
data_2024['ZIPCode'] = data_2024['ZIPCode'].astype(str)
```

```python
data_2024.head()
```

```python
data_2024.to_csv('bmore_police_2024.csv')
```
